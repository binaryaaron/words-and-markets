```python
import json, datetime
rawTweets = []
for line in open("../data/stocktweets.20141008-101028.json"):
    try: 
        rawTweets.append(json.loads(line))
    except:
        pass
    
```

```python
def date_hook(json_dict):
    for (key, value) in json_dict.items():
        try:
            json_dict[key] = datetime.datetime.strptime(value,'%a %b %d %H:%M:%S +0000 %Y')
        except:
            pass
    return json_dict
```

```python
datetweets = []
for line in open("../data/stocktweets.20141008-101028.json"):
    try: 
        datetweets.append(json.loads(line, object_hook=date_hook))
    except:
        pass
```
